Random variables can be classified into two main types:

1. Discrete Random Variables
A random variable is discrete if it takes on a countable number of distinct values. These values can be finite or countably infinite.

Characteristics:
The probability of each value is given by a probability mass function (PMF).

The sum of all probabilities must be 1.

Often associated with counting processes (e.g., number of heads in coin tosses, number of customers in a queue).

Examples:
The number of heads in 5 coin flips.

The number of students present in a classroom.

The number of emails received in a day.

2. Continuous Random Variables
A random variable is continuous if it takes on an uncountable number of values within an interval (e.g., any real number between 0 and 1).

Characteristics:
The probability of a specific single value is zero.

Described by a probability density function (PDF).

Probabilities are assigned over intervals using integration.

Typically associated with measurements (e.g., weight, height, time, temperature).

Examples:
The height of a randomly chosen person.

The time it takes for a bus to arrive.

The temperature at a given moment.

 Significance of Confidence Intervals in Statistical Analysis
Confidence intervals (CIs) are fundamental tools in statistical analysis that provide a range of plausible values for an unknown population parameter (like a mean or proportion) rather than just a single point estimate. They convey both the estimate's precision and the degree of uncertainty.

 Confidence Intervals Are Important beacause:
1. Quantifying Uncertainty
A point estimate (e.g., sample mean) alone doesn't indicate how accurate or reliable it is.

Confidence intervals provide a range of values within which the true parameter is likely to lie, along with a confidence level (e.g., 95%, 99%).

The width of the interval reflects the uncertainty or precision of the estimate. Narrower intervals indicate more precise estimates.

2. Interpretation of Results
Instead of claiming that the parameter is exactly a certain value, CIs allow statements like:

"We are 95% confident that the true mean lies between 72 and 78."

This approach acknowledges that any estimate derived from a sample has some degree of error.

3. Decision-Making & Hypothesis Testing
Confidence intervals are used to determine statistical significance.

If a confidence interval does not contain the null hypothesis value (e.g., 0 for a difference in means), the result is considered statistically significant.

CIs can also be used to compare groups or measure the effect size.

4. Comparing Estimates Across Studies
Confidence intervals are valuable for meta-analyses or systematic reviews, where results from multiple studies are combined.

Instead of relying solely on point estimates, overlapping or non-overlapping intervals help assess the consistency of findings across studies.

5. Assessing Sample Size Adequacy
Wider intervals indicate more variability or a smaller sample size.

Researchers can use CIs to determine if their sample size is sufficient for detecting meaningful effects or making accurate estimates.

6. Building Credibility & Transparency
Reporting confidence intervals rather than just p-values enhances credibility in scientific research.

Provides readers with more information about the reliability and precision of findings.

```python
## QUES 3) Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.
def bernoulli_pdf(x, p):
    """
    Calculate the Probability Distribution Function (PDF) of a Bernoulli distribution.

    Args:
        x (int): The random variable (0 or 1).
        p (float): The probability of success (0 <= p <= 1).

    Returns:
        float: Probability associated with the random variable x.
    """
    if x not in [0, 1]:
        raise ValueError("x must be either 0 or 1 for a Bernoulli distribution.")
    if not (0 <= p <= 1):
        raise ValueError("Probability p must be between 0 and 1.")

    # Calculate the PDF using the Bernoulli distribution formula
    probability = (p ** x) * ((1 - p) ** (1 - x))
    return probability


# Example usage
p = 0.7  # Probability of success
print(f"P(X=0): {bernoulli_pdf(0, p):.4f}")
print(f"P(X=1): {bernoulli_pdf(1, p):.4f}")
```

```python
## QUES 15) Simulate multiple samples from a normal distribution and verify the Central Limit Theorem.
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm, shapiro

# Setting the random seed for reproducibility
np.random.seed(42)

# Parameters for the original normal distribution
mu = 50       # Mean
sigma = 10    # Standard deviation
population_size = 10000  # Population size

# Generate a population from a normal distribution
population = np.random.normal(mu, sigma, population_size)

# Plotting the original population distribution
plt.figure(figsize=(12, 6))
sns.histplot(population, bins=50, kde=True, color='lightcoral')
plt.title('Original Population Distribution (Normal)', fontsize=16)
plt.xlabel('Value', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.grid(True)
plt.show()

# Implementing the Central Limit Theorem
num_sample_means = 1000  # Number of sample means to generate
sample_size = 30         # Size of each sample
sample_means = []

for _ in range(num_sample_means):
    sample = np.random.choice(population, sample_size, replace=True)
    sample_means.append(np.mean(sample))

# Plotting the distribution of sample means
plt.figure(figsize=(12, 6))
sns.histplot(sample_means, bins=50, kde=True, color='skyblue')
plt.title(f'Distribution of Sample Means (n={sample_size})', fontsize=16)
plt.xlabel('Sample Mean', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.grid(True)

# Plotting the theoretical normal distribution for comparison
x = np.linspace(min(sample_means), max(sample_means), 1000)
theoretical_pdf = norm.pdf(x, np.mean(sample_means), np.std(sample_means))
plt.plot(x, theoretical_pdf, 'r-', lw=2, label='Theoretical Normal Distribution')
plt.legend()
plt.show()

# Quantitative Normality Test - Shapiro-Wilk Test
stat, p_value = shapiro(sample_means)
print(f"Shapiro-Wilk Test: Statistic = {stat:.4f}, p-value = {p_value:.4f}")

if p_value > 0.05:
    print("The sample means distribution is likely normal (Fail to reject H0).")
else:
    print("The sample means distribution is likely not normal (Reject H0).")
```

```python
## QUES 19) Implement hypothesis testing using Z-statistics for a sample dataset.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def z_test(sample_mean, population_mean, population_std, sample_size, alpha=0.05):
    """
    Performs a one-sample Z-test and plots the distribution.

    Parameters:
    - sample_mean (float): The mean of the sample data.
    - population_mean (float): The known population mean.
    - population_std (float): The known population standard deviation.
    - sample_size (int): The number of observations in the sample.
    - alpha (float): The significance level for the hypothesis test.

    Returns:
    - z_score (float): The calculated Z-score.
    - p_value (float): The p-value corresponding to the Z-score.
    """
    # Calculate the standard error
    standard_error = population_std / np.sqrt(sample_size)

    # Calculate the Z-score
    z_score = (sample_mean - population_mean) / standard_error

    # Calculate the p-value for a two-tailed test
    p_value = 2 * (1 - norm.cdf(abs(z_score)))

    # Plotting the standard normal distribution
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x, 0, 1)

    plt.figure(figsize=(12, 6))
    plt.plot(x, y, label='Standard Normal Distribution', color='blue')
    plt.fill_between(x, y, where=(x <= -abs(z_score)) | (x >= abs(z_score)), color='orange', alpha=0.5, label='Rejection Region')
    plt.axvline(z_score, color='red', linestyle='--', label=f'Calculated Z-score = {z_score:.2f}')
    plt.axvline(-z_score, color='red', linestyle='--')

    plt.title('Hypothesis Testing Using Z-Statistics', fontsize=16)
    plt.xlabel('Z-score', fontsize=14)
    plt.ylabel('Probability Density', fontsize=14)
    plt.grid(True)
    plt.legend()
    plt.show()

    # Print the results
    print(f"Z-score: {z_score:.2f}")
    print(f"P-value: {p_value:.4f}")

    # Conclusion
    if p_value < alpha:
        print(f"Reject the null hypothesis (p-value < {alpha}).")
    else:
        print(f"Fail to reject the null hypothesis (p-value >= {alpha}).")

    return z_score, p_value


# Given data
population_mean = 70
population_std = 10
sample_mean = 74
sample_size = 30
alpha = 0.05

# Perform the Z-test
z_score, p_value = z_test(sample_mean, population_mean, population_std, sample_size, alpha)
```

```python
## QUES 20) Create a confidence interval for a dataset using Python and interpret the result.
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def confidence_interval(data, confidence=0.95):
    """
    Calculates the confidence interval for a given dataset.

    Parameters:
    - data (array-like): The dataset for which the confidence interval is calculated.
    - confidence (float): The desired confidence level (e.g., 0.95 for 95% confidence interval).

    Returns:
    - (float, float): The lower and upper bounds of the confidence interval.
    """
    # Calculate sample statistics
    n = len(data)
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)

    # Calculate standard error of the mean
    standard_error = sample_std / np.sqrt(n)

    # Calculate the critical Z-value
    z_value = stats.norm.ppf(1 - (1 - confidence) / 2)

    # Calculate the margin of error
    margin_of_error = z_value * standard_error

    # Calculate confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound


# Generating a random dataset (e.g., student scores)
np.random.seed(42)
data = np.random.normal(loc=70, scale=10, size=100)  # Mean = 70, Std = 10, Sample Size = 100

# Calculate the confidence interval
confidence = 0.95
lower, upper = confidence_interval(data, confidence)

# Print the result
print(f"Sample Mean: {np.mean(data):.2f}")
print(f"95% Confidence Interval: ({lower:.2f}, {upper:.2f})")

# Plotting the data and confidence interval
plt.figure(figsize=(12, 6))
plt.hist(data, bins=15, color='skyblue', edgecolor='black', alpha=0.7, density=True)
plt.axvline(np.mean(data), color='red', linestyle='--', label=f'Sample Mean = {np.mean(data):.2f}')
plt.axvline(lower, color='green', linestyle='--', label=f'Lower Bound = {lower:.2f}')
plt.axvline(upper, color='green', linestyle='--', label=f'Upper Bound = {upper:.2f}')
plt.title(f'{int(confidence * 100)}% Confidence Interval for the Dataset', fontsize=16)
plt.xlabel('Data Values', fontsize=14)
plt.ylabel('Probability Density', fontsize=14)
plt.legend()
plt.grid(True)
plt.show()
```

```python
## QUES 22) Write a Python script to calculate and visualize the probability density function (PDF) of a normal distribution.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_normal_pdf(mean, std_dev, x_range):
    """
    Plots the Probability Density Function (PDF) of a normal distribution.

    Parameters:
    - mean (float): The mean of the normal distribution.
    - std_dev (float): The standard deviation of the normal distribution.
    - x_range (tuple): The range of x-values to plot (min, max).
    """
    # Generate x values
    x = np.linspace(x_range[0], x_range[1], 1000)

    # Calculate the PDF using scipy's norm.pdf function
    pdf_values = norm.pdf(x, loc=mean, scale=std_dev)

    # Plotting the PDF
    plt.figure(figsize=(12, 6))
    plt.plot(x, pdf_values, color='blue', lw=2, label=f'N({mean}, {std_dev}^2)')
    plt.title('Probability Density Function (PDF) of a Normal Distribution', fontsize=16)
    plt.xlabel('x', fontsize=14)
    plt.ylabel('Probability Density', fontsize=14)
    plt.grid(True)
    plt.legend()
    plt.show()


# Parameters for the normal distribution
mean = 0
std_dev = 1
x_range = (-4, 4)

# Plotting the PDF
plot_normal_pdf(mean, std_dev, x_range)
```

```python
## QUES 23) Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

def plot_poisson_cdf(lam, max_k):
    """
    Plots the CDF of a Poisson distribution.

    Parameters:
    - lam (float): The average rate of occurrence (Î»).
    - max_k (int): The maximum value of k for plotting the CDF.
    """
    # Generate values for k
    k_values = np.arange(0, max_k + 1)

    # Calculate the CDF for each value of k using scipy's poisson.cdf()
    cdf_values = poisson.cdf(k_values, mu=lam)

    # Plotting the CDF
    plt.figure(figsize=(12, 6))
    plt.step(k_values, cdf_values, where='post', color='blue', lw=2, label=f'Î» = {lam}')
    plt.title(f'Cumulative Distribution Function (CDF) of a Poisson Distribution (Î» = {lam})', fontsize=16)
    plt.xlabel('k (Number of Events)', fontsize=14)
    plt.ylabel('Cumulative Probability', fontsize=14)
    plt.grid(True)
    plt.legend()
    plt.show()

    # Displaying the CDF values
    for k, cdf_val in zip(k_values, cdf_values):
        print(f"P(X <= {k}) = {cdf_val:.4f}")


# Parameters for the Poisson distribution
lam = 4  # Average rate of occurrence (Î»)
max_k = 15  # Maximum value of k to visualize

# Plotting the CDF
plot_poisson_cdf(lam, max_k)
```

```python
## QUES 24) Simulate a random variable using a continuous uniform distribution and calculate its expected value.
import numpy as np
import matplotlib.pyplot as plt

def simulate_continuous_uniform(a, b, n_samples):
    """
    Simulates a random variable using a continuous uniform distribution and calculates its expected value.

    Parameters:
    - a (float): Lower bound of the distribution.
    - b (float): Upper bound of the distribution.
    - n_samples (int): Number of samples to generate.

    Returns:
    - samples (numpy.ndarray): The generated samples.
    - expected_value (float): The calculated expected value of the distribution.
    """
    # Generate samples from a continuous uniform distribution
    samples = np.random.uniform(a, b, n_samples)

    # Calculate the theoretical expected value
    theoretical_expected_value = (a + b) / 2

    # Calculate the sample mean (approximation of the expected value)
    sample_mean = np.mean(samples)

    # Plotting the histogram of the simulated samples
    plt.figure(figsize=(12, 6))
    plt.hist(samples, bins=30, color='skyblue', edgecolor='black', density=True)
    plt.axvline(sample_mean, color='red', linestyle='--', label=f'Sample Mean = {sample_mean:.2f}')
    plt.axvline(theoretical_expected_value, color='green', linestyle='--', label=f'Theoretical Mean = {theoretical_expected_value:.2f}')
    plt.title(f'Continuous Uniform Distribution (a={a}, b={b})', fontsize=16)
    plt.xlabel('Value', fontsize=14)
    plt.ylabel('Density', fontsize=14)
    plt.legend()
    plt.grid(True)
    plt.show()

    return samples, sample_mean, theoretical_expected_value


# Parameters
a = 10   # Lower bound of the distribution
b = 20   # Upper bound of the distribution
n_samples = 10000  # Number of random samples to generate

# Simulate and calculate expected value
samples, sample_mean, theoretical_mean = simulate_continuous_uniform(a, b, n_samples)

# Displaying the results
print(f"Theoretical Expected Value: {theoretical_mean}")
print(f"Sample Mean (Approximation): {sample_mean}")
```

```python
## QUES 25) Write a Python program to compare the standard deviations of two datasets and visualize the difference
import numpy as np
import matplotlib.pyplot as plt

def compare_standard_deviations(data1, data2):
    """
    Compares the standard deviations of two datasets and visualizes the difference.

    Parameters:
    - data1 (numpy.ndarray): The first dataset.
    - data2 (numpy.ndarray): The second dataset.
    """
    # Calculate standard deviations
    std1 = np.std(data1, ddof=1)
    std2 = np.std(data2, ddof=1)

    # Print standard deviations
    print(f"Standard Deviation of Dataset 1: {std1:.4f}")
    print(f"Standard Deviation of Dataset 2: {std2:.4f}")

    # Plotting histograms for visualization
    plt.figure(figsize=(15, 6))

    plt.subplot(1, 2, 1)
    plt.hist(data1, bins=30, color='blue', alpha=0.7, label=f'Dataset 1 (Std = {std1:.2f})')
    plt.hist(data2, bins=30, color='orange', alpha=0.7, label=f'Dataset 2 (Std = {std2:.2f})')
    plt.title('Histograms of Two Datasets')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.legend()
    plt.grid(True)

    # Plotting boxplots for comparison
    plt.subplot(1, 2, 2)
    plt.boxplot([data1, data2], labels=['Dataset 1', 'Dataset 2'])
    plt.title('Box Plots of Two Datasets')
    plt.ylabel('Value')
    plt.grid(True)

    plt.tight_layout()
    plt.show()


# Generating two datasets with different standard deviations
np.random.seed(42)  # For reproducibility
data1 = np.random.normal(loc=50, scale=5, size=1000)   # Mean = 50, Std Dev = 5
data2 = np.random.normal(loc=50, scale=10, size=1000)  # Mean = 50, Std Dev = 10

# Compare and visualize
compare_standard_deviations(data1, data2)
```

```python
## QUES 26) Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution.
import numpy as np
import matplotlib.pyplot as plt

def calculate_range_and_iqr(data):
    """
    Calculates the range and interquartile range (IQR) of a dataset.

    Parameters:
    - data (numpy.ndarray): The dataset.

    Returns:
    - data_range (float): The range of the dataset.
    - iqr (float): The interquartile range (IQR) of the dataset.
    - q1 (float): The 25th percentile of the dataset.
    - q3 (float): The 75th percentile of the dataset.
    """
    # Calculate range
    data_range = np.max(data) - np.min(data)

    # Calculate Q1 (25th percentile) and Q3 (75th percentile)
    q1 = np.percentile(data, 25)
    q3 = np.percentile(data, 75)

    # Calculate IQR
    iqr = q3 - q1

    return data_range, iqr, q1, q3


# Generate a dataset from a normal distribution
np.random.seed(42)  # For reproducibility
data = np.random.normal(loc=50, scale=15, size=1000)  # Mean = 50, Std Dev = 15

# Calculate range and IQR
data_range, iqr, q1, q3 = calculate_range_and_iqr(data)

# Display results
print(f"Range of the dataset: {data_range:.2f}")
print(f"Interquartile Range (IQR) of the dataset: {iqr:.2f}")
print(f"Q1 (25th Percentile): {q1:.2f}")
print(f"Q3 (75th Percentile): {q3:.2f}")

# Plotting the distribution
plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram of Normally Distributed Data', fontsize=16)
plt.xlabel('Value', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.axvline(q1, color='orange', linestyle='--', label=f'Q1 = {q1:.2f}')
plt.axvline(q3, color='green', linestyle='--', label=f'Q3 = {q3:.2f}')
plt.legend()
plt.grid(True)
plt.show()
```

```python
## QUES 27) Implement Z-score normalization on a dataset and visualize its transformation.
import numpy as np
import matplotlib.pyplot as plt

def z_score_normalization(data):
    """
    Applies Z-score normalization to a dataset.

    Parameters:
    - data (numpy.ndarray): The original dataset.

    Returns:
    - normalized_data (numpy.ndarray): The Z-score normalized dataset.
    - original_mean (float): Mean of the original dataset.
    - original_std (float): Standard deviation of the original dataset.
    """
    original_mean = np.mean(data)
    original_std = np.std(data)

    # Applying Z-score normalization
    normalized_data = (data - original_mean) / original_std

    return normalized_data, original_mean, original_std


# Generate a dataset from a normal distribution
np.random.seed(42)
data = np.random.normal(loc=100, scale=20, size=1000)  # Mean = 100, Std Dev = 20

# Apply Z-score normalization
normalized_data, original_mean, original_std = z_score_normalization(data)

# Plotting the original and normalized datasets
plt.figure(figsize=(14, 6))

# Original Data Histogram
plt.subplot(1, 2, 1)
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title(f'Original Data Distribution (Mean={original_mean:.2f}, Std Dev={original_std:.2f})', fontsize=14)
plt.xlabel('Value', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(True)

# Normalized Data Histogram
plt.subplot(1, 2, 2)
plt.hist(normalized_data, bins=30, color='salmon', edgecolor='black')
plt.title('Z-Score Normalized Data Distribution (Mean=0, Std Dev=1)', fontsize=14)
plt.xlabel('Z-Score', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(True)

plt.tight_layout()
plt.show()

# Displaying transformation statistics
print(f"Original Mean: {original_mean:.2f}")
print(f"Original Standard Deviation: {original_std:.2f}")
print(f"Normalized Data Mean: {np.mean(normalized_data):.2f}")
print(f"Normalized Data Standard Deviation: {np.std(normalized_data):.2f}")
```

```python
## QUES 28)  Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal
## distribution.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis

def calculate_skewness_kurtosis(data):
    """
    Calculates the skewness and kurtosis of a dataset.

    Parameters:
    - data (numpy.ndarray): The dataset.

    Returns:
    - skewness (float): The skewness of the dataset.
    - kurt (float): The kurtosis of the dataset.
    """
    skewness = skew(data)
    kurt_value = kurtosis(data, fisher=True)  # Fisher=True gives excess kurtosis (Normal distribution = 0)

    return skewness, kurt_value


# Generate a dataset from a normal distribution
np.random.seed(42)
data = np.random.normal(loc=0, scale=1, size=1000)  # Mean = 0, Std Dev = 1

# Calculate skewness and kurtosis
skewness, kurt_value = calculate_skewness_kurtosis(data)

# Display results
print(f"Skewness of the dataset: {skewness:.4f}")
print(f"Kurtosis of the dataset: {kurt_value:.4f}")

# Plotting the distribution
plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram of Normally Distributed Data', fontsize=16)
plt.xlabel('Value', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.grid(True)
plt.show()
```
